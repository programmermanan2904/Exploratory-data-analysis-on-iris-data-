```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Load the Iris dataset
data = pd.read_csv('/content/1) iris.csv')
print(data.head())
```

```python
# 1.calculating summary statistics
# Mean, Median, Mode, Std for numerical columns
summary_stats = pd.DataFrame({
    "Mean": data.select_dtypes(include=np.number).mean(),
    "Median": data.select_dtypes(include=np.number).median(),
    "Mode": data.select_dtypes(include=np.number).mode().iloc[0],
    "Std Dev": data.select_dtypes(include=np.number).std()
})

print("Summary Statistics:")
print(summary_stats)
```

```python
# 2.data visualization
sns.set(style="whitegrid")

# a. histograms
data.hist(figsize=(10,6), bins=15, edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()
```

```python
# b. boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=data.select_dtypes(include=np.number))
plt.suptitle("boxplot of numerical values", fontsize=16)
plt.show()
```

```python
# c. scatterplot
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species')
plt.suptitle("scatterplot for sepal length and sepal width", fontsize=16)
plt.show()
```

```python
# 3.correlation analysis
corr_matrix = data.select_dtypes(include=np.number).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("correlation heatmap")
plt.show()
```
